### Drop unnecessary columns

```python
df.drop(['Timestamp','state','comments','Country'],axis=1,inplace=True)
```

- Drop Timestamp column because it's contains date, month, year and time the respondent took this questionnaire, which is irrelevant.
- Drop Country column because it's contained with the country where respondents live and 60% of answer in domination in 1 country. This column becomes irrelevant to the prediction model.
- Drop state column because it's only filled if the respondent lives in the US state, which means other respondents outside the US don't have any state ID and become missing values with 40% of data.
- Drop comments column because almost 70% of data is missing. It was an optional text box so it's reasonable to expect that many (most) respondents would leave it blank.

### Convert dtype object to int

```python
df['Age']=df['Age'].astype(int)
df['Gender'].replace({'Female':1,'Male':2,'Transgender':3},inplace=True)
df['self_employed'].replace({'Yes':1,'No':2},inplace=True)
df['family_history'].replace({'Yes':1,'No':2},inplace=True)
df['treatment'].replace({'Yes':0,'No':1},inplace=True)
df['work_interfere'].replace({'Sometimes':1,'Often':2,'Rarely':3,'Never':4},inplace=True)
df['no_employees'].replace({'01-May':1,'1-5':2,'6-25':3,'26-100':4,'100-500':5,'500-1000':6,'More than 1000':7,'Jun-25':8},inplace=True)
df['remote_work'].replace({'Yes':1,'No':2},inplace=True)
df['tech_company'].replace({'Yes':1,'No':2},inplace=True)
df['benefits'].replace({'Yes':1,'No':2,"Don't know":3},inplace=True)
df['care_options'].replace({'Yes':1,'No':2,"Not sure":3},inplace=True)
df['wellness_program'].replace({'Yes':1,'No':2,"Don't know":3},inplace=True)
df['seek_help'].replace({'Yes':1,'No':2,"Don't know":3},inplace=True)
df['anonymity'].replace({'Yes':1,'No':2,"Don't know":3},inplace=True)
df['leave'].replace({'Very difficult':1,'Somewhat difficult':2,'Very easy':3,'Somewhat easy':4,"Don't know":5},inplace=True)
df['mental_health_consequence'].replace({'Yes':1,'No':2,"Maybe":3},inplace=True)
df['phys_health_consequence'].replace({'Yes':1,'No':2,"Maybe":3},inplace=True)
df['coworkers'].replace({'Yes':1,'No':2,"Some of them":3},inplace=True)
df['supervisor'].replace({'Yes':1,'No':2,"Some of them":3},inplace=True)
df['mental_health_interview'].replace({'Yes':1,'No':2,"Maybe":3},inplace=True)
df['phys_health_interview'].replace({'Yes':1,'No':2,"Maybe":3},inplace=True)
df['mental_vs_physical'].replace({'Yes':1,'No':2,"Don't know":3},inplace=True)
df['obs_consequence'].replace({'Yes':1,'No':2},inplace=True)
```

## Creation of a Model

#### Take the X and y value

- Target: Treatment

```python
x=df.drop(['treatment'],axis=1)
```

```python
y=df['treatment']
```

### Split the Training Dataset and Test Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

## Model 1 : LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

Model1 = LogisticRegression(random_state=1)
Model1.fit(x_train,y_train)

y_pred1=Model1.predict(x_test)
print(classification_report(y_test, y_pred1))
```

- Overall Accuracy = 70%

## Model 2 : DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

Model2=DecisionTreeClassifier(random_state=1)
Model2.fit(x_train,y_train)

y_pred2=Model2.predict(x_test)
print(classification_report(y_test, y_pred2))
```

- Overall Accuracy = 76%

## Model 3 : Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

Model3=RandomForestClassifier(random_state=1)
Model3.fit(x_train,y_train)

y_pred3=Model3.predict(x_test)
print(classification_report(y_test, y_pred3))
```

- Overall Accuracy = 82%

## Model 4  : K-NN (K-Nearest Neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

Model4=KNeighborsClassifier()
Model4.fit(x_train,y_train)

y_pred4=Model4.predict(x_test)
print(classification_report(y_test, y_pred4))
```

- Overall Accuracy = 65%

## Model 5 : SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

Model5=SVC(random_state=1)
Model5.fit(x_train,y_train)

y_pred5=Model5.predict(x_test)
print(classification_report(y_test, y_pred5))
```

- Overall Accuracy = 64%

## Model 6 : Naives Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

Model6=GaussianNB()
Model6.fit(x_train,y_train)

y_pred6=Model6.predict(x_test)
print(classification_report(y_test, y_pred6))
```

- Overall Accuracy = 72%

## Model 7: XGBoost

```python
import xgboost as xgb
from xgboost import XGBClassifier

Model7 = XGBClassifier(random_state=1)
Model7.fit(x_train, y_train)

y_pred7 = Model7.predict(x_test)
print(classification_report(y_test, y_pred7))
```

- Overall Accuracy = 79%
